```python
quant_cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin']

for col in quant_cols:
    auto[col] = pd.to_numeric(auto[col], errors='coerce') 


for col in quant_cols:
    col_min = np.min(auto[col])
    col_max = np.max(auto[col])
    col_range = col_max - col_min
    print(f"{col}: min={col_min}, max={col_max}, range={col_range}")
```
